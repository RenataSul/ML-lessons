```python
import pandas as pd
```

```python
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import scale
```

1. Загрузите выборку

```python

data = pd.read_csv("wine.data", header=None)
data
```

2. Извлеките из данных признаки и классы. Класс записан в первом столбце (три варианта), признаки в столбцах со второго по последний.

```python
X = data.iloc[:, 1:]
y = data.iloc[:, 0]
X, y
```

3. Оценку качества необходимо провести методом кросс-валидации по 5 блокам (5-fold). Создайте генератор разбиений, который перемешивает выборку перед формированием блоков. Для воспроизводимости результата, создавайте генератор KFold с фиксированным параметром random_state=42. В качестве меры качества используйте долю верных ответов (accuracy).

```python
cross_val = KFold(n_splits = 5, shuffle = True, random_state=42)
cross_val
```

4. Найдите точность классификации на кросс-валидации для метода k ближайших соседей (sklearn.neighbors.KNeighborsClassifier), при k от 1 до 50. При каком k получилось оптимальное качество? Чему оно равно (число в интервале от 0 до 1)? Данные результаты и будут ответами на вопросы 1 и 2.

```python
scores = []
for k in range(1, 50+1):
  classif = KNeighborsClassifier(n_neighbors=k)
  scores.append(cross_val_score(classif, X, y, cv=cross_val, scoring='accuracy'))

print(pd.DataFrame(scores).mean().sort_values().index[-1] + 1, pd.DataFrame(scores).mean().sort_values().iloc[-1])
```

5. Произведите масштабирование признаков с помощью функции sklearn.preprocessing.scale. Снова найдите оптимальное k на кросс-валидации.

6. Какое значение k получилось оптимальным после приведения признаков к одному масштабу? Приведите ответы на вопросы 3 и 4. Помогло ли масштабирование признаков?

```python
X = scale(X)
scores = []
for k in range(1, 51):
    model = KNeighborsClassifier(k)
    scores.append(cross_val_score(model, X, y, cv = cross_val, scoring = 'accuracy'))
pd.DataFrame(scores, range(1, 51)).mean(axis=1).sort_values(ascending= False).head(1)
```
